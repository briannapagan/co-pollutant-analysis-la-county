This is geospatial data as designated by the latitude and longitude columns.
For the class, I wanted to understand what was the highest amount of co-pollutants in LA county.

```python
# group by polluant, and sum by emissions
total_emissions_by_pollutant = nei_facilities.groupby(['Pollutant','Pollutant Type'])['Emissions (Tons)'].sum().reset_index()
total_emissions_by_pollutant = total_emissions_by_pollutant.sort_values(by='Emissions (Tons)',ascending = False)
total_emissions_by_pollutant.head(20)
```

Some of the top 10 pollutants are GHG's, which are not considered to be co-pollutants. These include carbon dioxide, methane and nitrous oxide.
