This is geospatial data as designated by the latitude and longitude columns. We want to work with the geospatial version of pandas called geopandas or gpd for short. 


```python
# Convert the latitude and longitude to Point geometries
geometry = [Point(xy) for xy in zip(nei_facilities['Longitude'], nei_facilities['Latitude'])]

# Create a GeoDataFrame
gdf = gpd.GeoDataFrame(nei_facilities, geometry=geometry)
```
